```python
import csv # csv 파일 출력 모듈 임포트


class Aiffel:
    id_counter = 0 # 생성된 학번의 개수를 출력용 클래스 변수

    guru_dic = {
        "김건": 0, "김경민": 0, "김소연": 0, "김영신": 0, "박근수": 0,
        "박영수": 0, "박영준": 0, "서원영": 0, "신지만": 0, "심지안": 0,
        "이규상": 0, "이윤상": 0, "이진영": 0, "임현석": 0, "전요한": 0,
        "전재영": 0, "조대희": 0
    }

    def __init__(self, name, admission_year, classmate, course):
        self.name = name
        self.admission_year = admission_year
        self.classmate = classmate
        self.course = course
        self.student_number = self.generate_student_number()
        self.quest_score_total = 0
        self.quest_penalty_total = 0
        Aiffel.id_counter += 1

    def generate_student_number(self):
        year = str(self.admission_year)[-2:]
        class_num = str(self.classmate).zfill(2)

        if self.course == "리서치":
            course_num = "1"
        elif self.course == "코어":
            course_num = "2"
        else:
            course_num = "0"

        sorted_classmate = sorted(Aiffel.guru_dic.keys())
        index = str(sorted_classmate.index(self.name) + 1).zfill(2)

        student_number = year + class_num + course_num + index
        return student_number

    def update_student_number(self): # 학생별 학번 업데이트
        self.student_number = self.generate_student_number()
        Aiffel.guru_dic[self.name] = self.student_number

    def score(self, points): # 퀘스트 점수 누적
        if 1 <= points <= 3:
            self.quest_score_total += points

    def penalty(self, points): # 퀘스트 페널티 누적
        self.quest_penalty_total += points

    @classmethod
    def get_id_num(cls):
        return cls.id_counter

    def display_info(self): # Aiffel 인스턴스에 저장된 정보를 출력
        print("Name:", self.name)
        print("Student Number:", self.student_number)
        print("Course:", self.course)
        print("Quest Score Total:", self.quest_score_total)
        print("Quest Penalty Total:", self.quest_penalty_total)
        print("Quest Gross Total:", (self.quest_score_total - self.quest_penalty_total))

    """# CSV 파일로 출력
        with open('student_info.csv', 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(['Name', 'Student Number', 'Course', 'Quest Score Total', 'Quest Penalty Total'])
            writer.writerow([self.name, self.student_number, self.course, self.quest_score_total, self.quest_penalty_total])"""

    def drop_student(self, name):
        del Aiffel.guru_dic[self.name]

print(Aiffel.guru_dic)
print('=' * 30)

# student_number 업데이트
for name in Aiffel.guru_dic.keys():
    student = Aiffel(name, 2023, 25, '코어')
    student.update_student_number()
    student.score(2)
    student.penalty(1)
    student.score(2)
    student.penalty(2)
    student.display_info()

print('=' * 30)

print(Aiffel.guru_dic)

print('=' * 30)

print(Aiffel.get_id_num()) # Aiffel 클래스로부터 생성된 학번의 개수를 출력

print('=' * 30)

student = Aiffel("김경민", 2023, 25, "코어")  # 삭제할 학생 정보를 포함한 인스턴스 생성
student.drop_student(student.name)  # drop_student 메서드 호출하여 학생 정보 삭제

student = Aiffel("김영신", 2023, 25, "코어")  # 삭제할 학생 정보를 포함한 인스턴스 생성
student.drop_student(student.name)  # drop_student 메서드 호출하여 학생 정보 삭제

print('=' * 30)

print(Aiffel.guru_dic)
```
