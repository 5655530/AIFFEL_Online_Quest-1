```python
'''
Q1. Account 클래스 : 은행에 가서 계좌를 개설하면 은행이름, 예금주, 계좌번호, 잔액이 설정됩니다.
Account 클래스를 생성한 후 생성자(hint: 매직매서드..!!)를 구현해보세요.
생성자에서는 예금주와 초기 잔액만 입력 받습니다.
은행이름은 SC은행으로 계좌번호는 3자리-2자리-6자리 형태로 랜덤하게 생성됩니다. => class Account

Q2. 클래스 변수: 클래스 변수를 사용해서 Account 클래스로부터 생성된 계좌 객체의 개수를 저장하세요. => num_accounts

Q3. 클래스 변수 출력 : Account 클래스로부터 생성된 계좌의 개수를 출력하는 get_account_num() 메서드를 추가하세요. =>  def get_account_num()

Q4. 입금 메서드 : Account 클래스에 입금을 위한 deposit 메서드를 추가하세요. 입금은 최소 1원 이상만 가능합니다. => def deposit(self, amount)

Q5. 출금 메서드 : Account 클래스에 출금을 위한 withdraw 메서드를 추가하세요. 출금은 계좌의 잔고 이상으로 출금할 수는 없습니다. => def withdraw(self, amount)

Q6. 정보 출력 메서드 : Account 인스턴스에 저장된 정보를 출력하는 display_info() 메서드를 추가하세요. 잔고는 세자리마다 쉼표를 출력하세요. => def display_info(self)
(은행이름: SC은행, 예금주: 파이썬, 계좌번호: 111-11-111111, 잔고: 10,000원)

Q7. 이자 지급하기 입금 횟수가 5회가 될 때 잔고를 기준으로 1%의 이자가 잔고에 추가되도록 코드를 변경해보세요.  => def deposit(self, amount)

Q8. 여러 객체 생성 Account 클래스로부터 3개 이상 인스턴스를 생성하고 생성된 인스턴스를 리스트에 저장해보세요. => account_list = []

Q9. 객체 순회 반복문을 통해 리스트에 있는 객체를 순회하면서 잔고가 100만원 이상인 고객의 정보만 출력하세요. => for account in account_list:

MainQuest Q10. 입출금 내역
입금과 출금 내역이 기록되도록 코드를 업데이트 하세요.
입금 내역과 출금 내역을 출력하는 deposit_history와 withdraw_history 메서드를 추가하세요.
'''
import random # 계좌 번호 랜덤 생성용

class Account:
    num_accounts = 0  # 클래스 변수, 모든 클래스 공통으로 계좌 생성 건수 추적

    def __init__(self, owner, balance): # 예금주, 잔액 매개변수
        self.bank = "SC은행" # 은행 이름
        self.owner = owner
        self.account_number = self.generate_account_number() # 계좌번호 자동 생성
        self.balance = balance # 초기 잔액 설정
        self.deposit_count = 0  # 입금 횟수 카운트 변수
        self.transaction_history = []  # 입출금 내역을 저장할 리스트
        self.transaction_history.append(f"입금: +{balance:,}원") # 초기 잔액을 입금 내역에 기록
        Account.num_accounts += 1  # 클래스 변수 값 증가

    def generate_account_number(self): # 계좌 번호 랜덤 생성 메소드
        random.seed() # 초기화
        account_number = f"{random.randint(100, 999)}-{random.randint(10, 99)}-{random.randint(100000, 999999)}" # ₩d{3}-\d{2}-\d{6} 포맷의 계좌 번호 새성
        return account_number

    @staticmethod # 클래를 변수 직접 호출
    def get_account_num():
        return Account.num_accounts

    def deposit(self, amount): # 1만원 이상 입금 메소드
        if amount >= 10000:
            self.balance += amount # 잔고에 입금액 합산
            self.transaction_history.append(f"입금: +{amount:,}원") # 입금 내역 기록
            self.deposit_count += 1 # 입금 횟수 카운트
            print(f'{self.owner}님, {amount:,}원이 입금되었습니다.')
            if self.deposit_count % 5 == 0: # 5회 입금 시 마다
                interest = self.balance * 0.01  # 잔고의 1% 이자 계산
                self.balance += interest # 이자 지급
                print(f'이자 {interest:,}원이 추가 입금되었습니다.')
                print(f'현재 잔고는 {self.balance:,}원입니다.')
        else:
            print("최소 1만원 이상의 금액을 입금해야 합니다.")

    def withdraw(self, amount): # 인출 메소드
        if amount > 0 and amount <= self.balance: # 잔고와 인출 금액 비교
            self.balance -= amount
            self.transaction_history.append(f"출금: -{amount:,}원") # 출금 내역 기록
            print(f'{self.owner}님, {amount:,}원이 출금되었습니다.')
            print(f'{self.owner}님, 잔고는 {self.balance:,}원입니다.')
        elif amount > self.balance:
            print("잔고 부족으로 요청하신 금액을 출금할 수 없습니다.")
        else:
            print("유효한 출금 금액을 입력해야 합니다.")

    def display_info(self): # ccount 인스턴스에 저장된 정보 출력 메소드
        print(f"은행이름: {self.bank}")
        print(f"예금주: {self.owner}")
        print(f"계좌번호: {self.account_number}")
        print(f"잔고: {self.balance:,}원")

    def deposit_history(self): # 입금 내역 출력 메소드
        print(f"{self.owner}님의 입금 내역:")
        for transaction in self.transaction_history:
            if transaction.startswith("입금"):
                print(transaction)

    def withdraw_history(self): # 출금 내역 출력 메소드
        print(f"{self.owner}님의 출금 내역:")
        for transaction in self.transaction_history:
            if transaction.startswith("출금"):
                print(transaction)


account_list = [] # 생선된 인스턴스를 저장할 빈 리스트 생성

my_account = Account("신지만", 10000000) # my_account 인스턴스 생성
account_list.append(my_account) # account_list에 인스턴스 추가

guru1_account = Account("전재영", 10000000) # guru1_account 인스턴스 생성
account_list.append(guru1_account)

guru2_account = Account("박영수", 10000000) # guru2_account 인스턴스 생성
account_list.append(guru2_account)

guru3_account = Account("심지안", 10000000) # guru3_account 인스턴스 생성
account_list.append(guru3_account)

'''
print('=' * 30)
my_account.deposit(1000) # 일천원 입급
my_account.deposit(1000000) # 일백만원 입급
print('=' * 30)
my_account.withdraw(1000) # 일천원 출금
my_account.withdraw(100000) # 일십만원 출금
print('=' * 30)
my_account.display_info()
print('=' * 30)
'''
# 이자 지급 및 거래 내역 테스트
for i in range(5):

    deposit_money = random.randint(10000, 15000)
    my_account.deposit(deposit_money)

    withdraw_money = random.randint(10000, 15000)
    my_account.withdraw(withdraw_money)
print('@' * 30)


for account in account_list: # 각각의 인스턴스 객체를 호출해 정보 출력
    if account.balance >= 1000000:
        account.display_info()
        print('*' * 20)

# my_account 입금 내역 출력
print()
my_account.deposit_history()
print('=' * 30)
# my_account 출금 내역 출력
my_account.withdraw_history()
print('=' * 30)
```
