```python
# 문제 2-1. Kiosk 클래스를 생성합니다.
import re
class Kiosk:
    def __init__(self):
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]

    def menu_print(self): # 메뉴판 출력하기
      for i in range(len(self.menu)): # 메뉴의 개수만큼 반복해서
        print(f'({i+1}) {self.menu[i]} : {self.price[i]}원') # 메뉴와 가격을 출력합니다.

    # 주문 메서드
    def menu_select(self):
        print() # 줄 바꿈
        self.order_menu = [] # 주문 리스트
        self.order_price = [] # 가격 리스트

        n = 0 # 음료 종류 선택
        while n < 1 or len(self.menu) < n:
            n = int(re.sub(r'[^0-9]', '', input("음료 번호를 입력하세요 : ")))  # 음료 번호 입력

            # 메뉴판에 있는 음료 번호일 때
            if 1 <= n & n <= len(self.menu):
                self.order_price.append(self.price[n-1])  # 가격 리스트에 추가합니다.
                self.price_sum = self.price[n-1]  # 합계 금액

            # 메뉴판에 없는 번호일 때
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        # 음료 온도 물어보기
        t = 0  # 기본값을 넣어주고
        while ((t != 1) and (t != 2)):  # 1이나 2를 입력할 때까지 물어보기
            t = int(re.sub(r'[^0-9]', '', input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : ")))
            if (t == 1):
              self.temp = 'HOT'
            elif (t == 2):
              self.temp = 'ICE'
            else:
              print("1과 2 중 하나를 입력하세요.\n")
            #print({self.temp})
        self.order_menu.append(self.temp + ' ' + self.menu[n-1])  # 주문 리스트에 추가합니다.

        # 문제 2-2의 답을 입력하세요.
        print(f'주문 음료 => {self.temp} {self.menu[n-1]} : {self.price[n-1]}원')  # 온도 속성을 추가한 주문 결과 출력


        # 추가 주문 또는 지불
        while n != 0:  # 지불을 선택하기 전까지 반복합니다.
            print()  # 줄 바꾸면서
            n = int(re.sub(r'[^0-9]', '', input("추가 주문은 음료 번호(1~6)를, 결제를 원하면 0을 누르세요 : ")))  # 추가 주문 또는 지불
            if n > 0 and n <= len(self.menu):

                # 추가 음료 온도
                # 문제 2-4. 추가 음료의 온도를 입력받고, 가격 리스트, 주문 리스트, 합계 금액을 업데이트해보세요.
                t = 0  # 기본값을 넣어주고
                while ((t != 1) and (t != 2)):  # 1이나 2를 입력할 때까지 물어보기
                    t = int(re.sub(r'[^0-9]', '', input("음료 온도 선택 => HOT 음료는 1을, ICE 음료는 2를 입력하세요 : ")))
                    if (t == 1):
                        self.temp = 'HOT'
                    elif (t == 2):
                        self.temp = 'ICE'
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")

                self.order_price.append(self.price[n-1])  # 가격 리스트에 추가합니다.
                self.price_sum += self.price[n-1]  # 합계 금액
                self.order_menu.append(self.temp + ' ' + self.menu[n-1])  # 주문 리스트에 추가합니다.

                print(f'추가 주문 음료 => {self.temp} {self.menu[n-1]} : {self.price[n-1]}원\n 합계 : [{self.price_sum}]원')
            else:
                if n == 0: # 지불을 입력하면
                    print("주문이 완료되었습니다.")
                    print(self.order_menu, self.order_price)  # 확인을 위한 리스트를 출력합니다.
                else:  # 없는 번호를 입력할 때
                    print("없는 메뉴입니다. 다시 주문해 주세요.")
    # 지불 방식 물어보기
    def pay(self):
        while True:  # 정상적인 지불 방식을 선택하기 전까지 반복
            print()  # 줄 바꾸면서
            n = int(re.sub(r'[^0-9]', '', input("지불 방식 선택 => 현금이면 1을 카드면 2를 입력하세요: "))) # 결제 방식 선택 입력
            if n == 1:
                print('직원을 호출하겠습니다.')
                break
            elif n == 2:
                print('IC칩 방향에 맞게 카드를 꽂아주세요.')
                break
            else:
                print('다시 결제를 시도해 주세요.')


    # 주문 내역 출력하기
    def table(self):
        # 외곽
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')

        # 주문 내역과 총 금액 출력
        for i in zip(self.order_menu, self.order_price):
            print(f'{i[0]} : {i[1]}')
        print(f'합계 : {self.price_sum}')

        # 외곽
        for i in range(5):
            print('|' + ' ' * 31+ '|')
        print('⟝' + '-' * 30 + '⟞')

a = Kiosk() # 인스턴스 생성
#print(a.menu)
#print(a.price)
a.menu_print() # 인스턴스의 메뉴 출력 메소드 호출

a.menu_select() # 인스턴스의 메뉴 선택 메소드 호출

a.pay() # 인스턴스의 지불 방식 선택 메소드 호출

a.table() # 인스턴스의 주문 내역 출력 메소드 호출

```
