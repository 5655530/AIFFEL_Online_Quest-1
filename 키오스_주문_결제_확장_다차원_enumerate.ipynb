```python
import re

class Kiosk:
    def __init__(self):
        self.menu_price = [['americano', 2000], ['latte', 3000], ['mocha', 3000], ['yuza_tea', 2500], ['green_tea', 2500], ['choco_latte', 3000]] # 이차원 리스트

    def menu_print(self): # 메뉴판 출력하기
        for i, item in enumerate(self.menu_price): # enumerate(), 메뉴의 개수만큼 반복해서
            menu, price = item # 언팩킹
            print(f'({i+1}) {menu} : {price}원') # 메뉴와 가격을 출력합니다.

    def menu_select(self):
        print()
        self.order_menu = [] # 주문 리스트
        self.order_price = [] # 가격 리스트
        self.price_sum = 0 # 주문 총금액 초기값

        n = 0 # 음료 종류 선택
        while n < 1 or len(self.menu_price) < n: # 1 ~ 6이 입력되면 종료
            n = int(re.sub(r'[^0-9]', '', input("음료 번호를 입력하세요: "))) # 음료 번호 입력

            # 메뉴판에 있는 음료 번호일 때
            if 1 <= n <= len(self.menu_price):
                menu, price = self.menu_price[n-1]
                self.order_price.append(price) # 가격 리스트에 추가합니다.
                self.price_sum = price # 해당 메뉴의 가격을 합계 금액에 할당

            # 메뉴판에 없는 번호일 때
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        # 음료 온도 물어보기
        t = 0 # 초기값을 넣어주고
        while t != 1 and t != 2: # 1이나 2를 입력할 때까지 물어보기
            t = int(re.sub(r'[^0-9]', '', input("HOT 음료는 1을, ICE 음료는 2를 입력하세요: ")))
            if t == 1:
                self.temp = 'HOT'
            elif t == 2:
                self.temp = 'ICE'
            else:
                print("1과 2 중 하나를 입력하세요.\n")

        menu, price = self.menu_price[n-1]
        self.order_menu.append(self.temp + ' ' + menu) # 주문 리스트에 추가합니다.
        print(f'주문 음료 => {self.temp} {menu} : {price}원') # 온도 속성을 추가한 주문 결과 출력

        # 추가 주문 또는 지불
        m = 1 # 추가 주문 또는 결제 선택
        while m != 0: # 지불을 선택하기 전까지 반복합니다.
            print()
            m = int(re.sub(r'[^0-9]', '', input("추가 주문은 음료 번호(1~6)를, 결제를 원하면 0을 누르세요: "))) # 추가 주문 또는 지불

            if m > 0 and m <= len(self.menu_price):
                # 추가 음료 온도
                t = 0
                while t != 1 and t != 2: # 1이나 2를 입력할 때까지 물어보기
                    t = int(re.sub(r'[^0-9]', '', input("음료 온도 선택 => HOT 음료는 1을, ICE 음료는 2를 입력하세요: ")))
                    if t == 1:
                        self.temp = 'HOT'
                    elif t == 2:
                        self.temp = 'ICE'
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")

                menu, price = self.menu_price[m-1]
                self.order_price.append(price) # 가격 리스트에 추가합니다.
                self.price_sum += price # 합계 금액 누적
                self.order_menu.append(self.temp + ' ' + menu) # 주문 리스트에 추가합니다.
                print(f'추가 주문 음료 => {self.temp} {menu} : {price}원\n 합계 : [{self.price_sum}원]')
            else:
                if m == 0: # 지불을 입력하면
                    print("주문이 완료되었습니다.")
                    print(self.order_menu, '\n', self.order_price) # 확인을 위한 주문, 가격 리스트를 출력합니다.
                    break
                else: # 없는 번호를 입력할 때
                    print("없는 메뉴입니다. 다시 주문해 주세요.")

    def pay(self):
        while True: # 정상적인 지불 방식을 선택하기 전까지 반복
            print()
            n = int(re.sub(r'[^0-9]', '', input("지불 방식 선택 => 현금이면 1을 카드면 2를 입력하세요: "))) # 결제 방식 선택 입력
            if n == 1:
                print('직원을 호출하겠습니다.')
                break
            elif n == 2:
                print('IC칩 방향에 맞게 카드를 꽂아주세요.')
                break
            else:
                print('다시 결제를 시도해 주세요.')

    # 주문 내역 출력하기
    def table(self):
        print()
        # 외곽
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')

        # 주문 내역과 총 금액 출력
        for item in zip(self.order_menu, self.order_price):
            menu, price = item
            print(f'{menu} : {price}')

        print(f'합계 : {self.price_sum}')

        # 외곽
        for i in range(5):
            print('|' + ' ' * 31+ '|')
        print('⟝' + '-' * 30 + '⟞')

a = Kiosk()
a.menu_print()
a.menu_select()
a.pay()
a.table()
```
