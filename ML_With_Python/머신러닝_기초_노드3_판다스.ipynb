```python
# apply 함수 => 데이터 전처리에 유용
# 데이터 확인
display(df)
def cal(x):
    if x >= 1300:
        return 'Yes'
    else:
        return 'No'

df['칼로리'].apply(cal)
```

```python
# apply 함수를 적용해서 새로운 컬럼을 생성
df['살찔까요'] = df['칼로리'].apply(cal)
display(df)
```

```python
# lambda, apply 활용
df['고민'] = df['칼로리'].apply(lambda x: '먹지말자' if x >= 1300 else '무조건먹자')
display(df)
```

```python
# Q. 다음 df 변수의 데이터에서 '가성비'라는 칼럼을 만들고,
# '호수' / '가격' 의 값이 0.0008을 넘으면 '가성비' 칼럼 값을 'Good' 아니면 'Bad'로 넣어주세요.
# 힌트 : astype(), apply, lambda를 활용합니다.
df = pd.read_csv('new_modudak.csv')
df['호수'] = df['호수'].str.replace('호', '')

df['호수'] = df['호수'].astype(int)
df
#df.info()
df['가성비'] = df[['호수', '가격']].apply(lambda x: 'Good' if (x['호수'] / x['가격']).astype(float) >= 0.0008 else 'Bad', axis=1)
df
```
