```python
# 다중 컬럼 수동 레이블 인코딩
df = data[['가격', '호수', '칼로리', '원산지', '살찔까요']].copy()
from sklearn.preprocessing import LabelEncoder
cols = ['원산지', '살찔까요'] # 인코딩 컬럼 리스트 변수

for col in cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col]) # 컬럼 단위 업데이트

display(df)
```

```python
# 다중 컬럼 자동 레이블 인코딩
df = data[['가격', '호수', '칼로리', '원산지', '살찔까요']].copy()
from sklearn.preprocessing import LabelEncoder # 사이킷런 모듈에서 레이블인코더 메소드 임포트
cols = df.select_dtypes(include='object').columns # object 타입의 모든 컬럼을 리스트로 추출 후 변수에 저장

for col in cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col]) # 컬럼 단위 업데이트

display(df)
```

```python
# 단일 컬럼 수동 원-핫 인코딩
df = data[['가격', '호수', '칼로리', '원산지', '살찔까요']].copy()
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False)
cat = ohe.fit_transform(df[['원산지']]) # 단일 컬럼명을 데이터프레임 형식으로 지정
cat
#ohe.categories_ # 컬럼명 확인 방법1
#ohe.get_feature_names_out() # 컬럼명 확인 방법2, 권장 방법, 컬럼명 생성
df_cat = pd.DataFrame(cat, columns=ohe.get_feature_names_out()) # 원-핫 인코딩 "데이터프레임" 생성 => concat()로 합치기 위해
display(df_cat)
df = pd.concat([df, df_cat], axis=1) # 열기준으로 두 데이터프레임 합치기
df

df = df.drop(['원산지'], axis=1) # 불필요한 object 컬럼 삭제
df
```

```python
# 다중 컬럼 수동 원-핫 인코딩
df = data[['가격', '호수', '칼로리', '원산지', '살찔까요']].copy()
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False)
cat = ohe.fit_transform(df[['원산지', '살찔까요']]) # 다중 컬럼명을 데이터프레임 형식으로 지정
cat
#ohe.categories_ # 컬럼명 확인 방법1
#ohe.get_feature_names_out() # 컬럼명 확인 방법2, 권장 방법, 컬럼명 생성
df_cat = pd.DataFrame(cat, columns=ohe.get_feature_names_out()) # 원-핫 인코딩 "데이터프레임" 생성 => concat()로 합치기 위해
display(df_cat)
df = pd.concat([df, df_cat], axis=1) # 열기준으로 두 데이터프레임 합치기
df

df = df.drop(['원산지', '살찔까요'], axis=1) # 불필요한 object 컬럼 삭제
df
```

```python
# 다중 컬럼 자동 원핫인코딩
df = data[['가격', '호수', '칼로리', '원산지', '살찔까요']].copy()
from sklearn.preprocessing import OneHotEncoder
cols = df.select_dtypes(include='object').columns # object 타입의 모든 컬럼을 리스트로 추출

ohe = OneHotEncoder(sparse_output=False)
cat = ohe.fit_transform(df[cols]) # 원핫인코딩 결과 값
#cat
#ohe.get_feature_names_out() # 컬럼명 생성 => "원산지_국내산, 원산지_미국, 원산지_브라질, 살찔까요_no,살찔까요_yes"
df_cat = pd.DataFrame(cat, columns=ohe.get_feature_names_out()) # 원-핫 인코딩 "데이터프레임" 생성 => concat()로 합치기 위해
df_cat

df = pd.concat([df, df_cat], axis=1) # 열기준으로 두 데이터프레임 합치기
df.head()

df = df.drop(cols, axis=1) # 불필요한 object 컬럼 삭제
df
```
