```python
# 다중 컬럼 수동 레이블인코딩
df = data[['가격', '호수', '칼로리', '원산지', '살찔까요']].copy()
#display(df)
#df.info()
cols = ['원산지', '살찔까요']
from sklearn.preprocessing import LabelEncoder #

for col in cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

display(df)
```

```python
# 다중 컬럼 자동 레이블인코딩
df = data[['가격', '호수', '칼로리', '원산지', '살찔까요']].copy()
#display(df)
#df.info()
cols = df.select_dtypes(include='object').columns # object 타입의 컬럼만 리스트로 추출
from sklearn.preprocessing import LabelEncoder # 사이킷런 모듈에서 레이블인코더 메소드 임포트

for col in cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

display(df)
```

```python
# 단일 컬럼 원핫인코딩
df = data[['가격', '호수', '칼로리', '원산지', '살찔까요']].copy()
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output=False)
cat = ohe.fit_transform(df[['원산지']])
cat
ohe.categories_
ohe.get_feature_names_out() # 권장 방법
df_cat = pd.DataFrame(cat, columns=ohe.get_feature_names_out())

df = pd.concat([df, df_cat], axis=1)
df

df = df.drop(['원산지'], axis=1)
df
```

```python
# 다중 컬럼 원핫인코딩
df = data[['가격', '호수', '칼로리', '원산지', '살찔까요']].copy()
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output=False)
cat = ohe.fit_transform(df[['원산지', '살찔까요']])
cat
#ohe.categories_
ohe.get_feature_names_out() # 권장 방법
df_cat = pd.DataFrame(cat, columns=ohe.get_feature_names_out())

df = pd.concat([df, df_cat], axis=1)
df

df = df.drop(['원산지', '살찔까요'], axis=1)
df
```
