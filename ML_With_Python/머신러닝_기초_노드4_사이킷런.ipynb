```python
# 사이킷런 데이터셋 불러오기
from sklearn.datasets import load_breast_cancer
dataset = load_breast_cancer()
#display(dataset)

# 피처(컬럼) 이름 확인
dataset.feature_names

# 타겟(레이블, 학습 데이터) 확인
dataset.target

# 데이터 확인
dataset.data[:2]

# 데이터프레임으로 만들기
cancer_df = pd.DataFrame(data=dataset.data, columns=dataset.feature_names)
#display(cancer_df)

# 타겟 추가하기
#display(dataset.target)
cancer_df['target'] = dataset.target
cancer_df.head()

# 사이킷런에서 제공하는 데이터셋 확인 (load_*)
import sklearn.datasets
sklearn.datasets.__all__
```

```python
# 당료병 데이터 불러오기
from sklearn.datasets import load_diabetes
dataset = load_diabetes()
#display(dataset)

# 피처(컬럼) 이름 확인 하기
dataset.feature_names

# 타겟(레이블, 학습 데이터) 확인
dataset.target

# 데이터프레임으로 만들기
diabetes_df = pd.DataFrame(data=dataset.data, columns=dataset.feature_names)
#display(diabetes_df)


# 타겟 추가하기
diabetes_df['target'] = dataset.target
diabetes_df.head()
```

```python
# 머신러닝 분류
# 검증(validation) 데이터 분리
# X_train, y_train 데이터를 이용해 머신러닝 훈련 & 학습
# X_test을 통해 예측
# y_test로 평가/검증
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    cancer_df.drop('target', axis=1), cancer_df['target'], test_size=0.3, random_state=1004)

# 훈련(training) 데이터 확인: (X) => 훈련 데이터(X_train)와 예측 데이터(X_test)
X_train.head()

# 학습(learning) 데이터 확인: 레이블(y) => 학습 데이터(y_train)와 검증 데이터(y_test)
y_train.head()

# 데이터 크기
X_train.shape, X_test.shape, y_train.shape, y_test.shape
# 여기까지가 학습과 평가에 필요한 데이터를 분리하는 작업
```

```python
# 1. 머신러닝: 의사결정나무(Decision Tree)
# X_train, y_train 데이터를 이용해 머신러닝 훈련 & 학습
# X_test을 통해 예측
# y_test로 평가

# 머신러닝 분류
# 의사결정트리 모델
from sklearn.tree import DecisionTreeClassifier

# 모델 학습
model = DecisionTreeClassifier()

# 훈련 & 학습, X_train, y_train 데이터를 이용해 머신러닝 훈련 & 학습
model.fit(X_train, y_train)

# 예측값, X_test을 통해 예측
pred = model.predict(X_test)
display(pred)

# 평가(Accuracy), y_test로 평가
# 정확도 accuracy_score(실제값, 예측값)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

```python
# 2. 머신러닝: 회귀분석
# 검증 데이터 분리
# X_train, y_train 데이터를 이용해 머신러닝 훈련 & 학습
# X_test을 통해 예측
# y_test로 평가
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    diabetes_df.drop('target', axis=1), diabetes_df['target'], test_size=0.3, random_state=1004)

# 훈련 데이터 확인
X_train.head()

# 학습 데이터 확인
y_train.head()

# 데이터 크기
X_train.shape, X_test.shape, y_train.shape, y_test.shape
# 여기까지가 학습과 평가에 필요한 데이터를 분리하는 작업
```

```python
# 선형회귀(Linear Regression)
# 머신러닝(회귀)
from sklearn.linear_model import LinearRegression

# 모델 학습
model = LinearRegression()

# 훈련 & 학습, X_train, y_train 데이터를 이용해 머신러닝 훈련 & 학습
model.fit(X_train, y_train)

# 예측값, X_test을 통해 예측
pred = model.predict(X_test)
display(pred)
```

```python
# 평가(mse)
# 정확도 mean_squared_error (실제값, 예측값)
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, pred)
```
