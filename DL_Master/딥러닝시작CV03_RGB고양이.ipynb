```python
# 필요한 모듈 불러오기
import numpy as np
from scipy.signal import convolve2d # 2D convolution 연산
from skimage.io import imread, imshow # 데이터를 이미지로 보여주기
from skimage.color import rgb2gray # RGB 채널값을 가진 이미지 데이터를, 흑백 이미지의 데이터로 변환하기
```

```python
cat = rgb2gray(imread('data/cat.jpg')) # read image
```

```python
# 흑백 고양이 사진은 3,266개의 행과 4,899개의 열로 이루어진 1-channel 이미지입니다.
# 행렬 안에는 0과 1 사이의 값으로 정규화된 픽셀 값이 들어 있습니다.
```

```python
cat.shape
```

```python
print(cat)
```

```python
imshow(cat)
```

```python
# Edge Detection
edge = np.array([[-1, -1, -1],
                    [-1, 8, -1],
                    [-1, -1, -1]])
# Sharpen
sharpen = np.array([[0, -1, 0],
                    [-1, 5, -1],
                    [0, -1, 0]])
# Gaussian Blur
blur = (4 / 16.0) * np.array([[1., 2., 1.],
                                 [2., 4., 2.],
                                 [1., 2., 1.]])
```

```python
edge_im = convolve2d(cat, edge, 'valid')
```

```python
imshow(edge_im)
```

```python
sharp_im = convolve2d(cat, sharpen, 'valid')
```

```python
# Q. Sharpen filter가 적용된 이미지를 출력해 보세요.
imshow(sharp_im)
```

```python
# Q. Gaussian Blur filter를 이미지에 적용해 보세요.
blur_im = convolve2d(cat, blur, 'valid')
```

```python
# Q. Gaussian Blur filter가 적용된 이미지를 출력해 보세요.
imshow(blur_im)
```

```python
# 이미지 불러오기
doggy = rgb2gray(imread('./data/doggy.jpg')) # read image
doggy.shape
imshow(doggy)
```

```python
# Edge Detection 적용
edge_im = convolve2d(doggy, edge, 'valid')
imshow(edge_im)
```

```python
# Gaussian Blur 적용
blur_im = convolve2d(doggy, blur, 'valid')
imshow(blur_im)
```

```python
# Sharpen 적용
sharpen_im = convolve2d(doggy, sharpen, 'valid')
imshow(sharpen_im)
```
