```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 준비
x = np.arange(0, 6, 0.1) # 0에서 6까지 0.1 간격으로 생성
x
y1 = np.sin(x) # 사인 함수
y2 = np.cos(x) # 코사인 함수

# 그래프 그리기
plt.plot(x, y1, label="sin")
plt.plot(x, y2, label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
